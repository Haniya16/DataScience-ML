# Salary Prediction
## Machine Learning Model
### Apply Linear Regression
---

```python
# import some libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
```

## Import the csv file

```python
# Attach the dataset file
df = pd.read_csv("salary_dataset.csv")
# check the dataset
df.head()
```

## Analyze Data

```python
#Unnamed: 0 column is same as like index column we can drop that column
df.drop(columns="Unnamed: 0",axis = 1,inplace = True)
df.tail()
```

```python
#null values
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Data distribution

plt.figure(figsize=(6,4))
plt.title('Salary Distribution Plot')
sns.histplot(df["Salary"], kde=True, stat="count", kde_kws=dict(cut=3))
plt.show()
```

## Relation Between Years of Experience & Salary

```python
# scatter plot before apply test_train the model
# relation between salary and experience

plt.figure(figsize=(6,4))
sns.scatterplot(data=df,x="YearsExperience",y="Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

# Split Data
### Split Data into Dependent & Independents

```python
# Splitting variables
# independent
X = df.iloc[:, :1]
# dependent
y = df.iloc[:, 1:]
```

## Split for Train & Test
---

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
#sizes of test and train data
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

## Train Model

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predict result

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

# Visualization Prediction
## Prediction on training set
---

```python
# Prediction on training set

plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

## Prediction on testing set
---

```python
# Prediction on test set
plt.figure(figsize=(6,4))
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

## Coefficient & Intercept
---

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
